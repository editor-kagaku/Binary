```python
def filetype_check(filepath):
    signature = [
        # 説明,オフセットアドレス,シグネチャ
        ["Executable and Linkable Format",0,b"\x7F\x45\x4C\x46"],
        ["PDF document",0,b"\x25\x50\x44\x46\x2D"],
    ]
    with open(filepath, "rb") as f:
        # 読み込むサイズはシグネチャリスト次第で調整
        head = f.read(12)
        for s in signature:
            if head[s[1]:s[1]+len(s[2])] == s[2]:
                return s[0]
    return "Unknown"
```

```python
filetype_check("sample_data/README.md")
```

```python
filetype_check("/usr/bin/ls")
```
