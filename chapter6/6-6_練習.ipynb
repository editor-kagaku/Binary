```python
%%file hello.c
#include <stdio.h>

char *rot13(char *s) {
    int i;
    char storeh[] = "NOPQRSTUVWXYZABCDEFGHIJKLM";
    char storel[] = "nopqrstuvwxyzabcdefghijklm";

    for (i = 0; s[i] != '\0'; i++) {
        if ((s[i] > 64 && s[i] < 91) || (s[i] > 96 && s[i] < 123)) {
            s[i] = (s[i] - 65 > 25) ? storel[s[i] - 97] : storeh[s[i] - 65];
        }
    }
    return (s);
}

int main(void) {
    char s[] = "Hello, World!";
    puts(rot13(s));
    puts(rot13(s));
    return 0;
}
```
