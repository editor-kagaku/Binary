```python
def c2a(c):
    if c >= 0x21 and c <=0x7E:
        return chr(c)
    elif c <= 0x20:
        return ["NUL","SOH","STX","ETX","EOT","ENQ","ACK","BEL",
                "BS","HT","LF","VT","FF","CR","SO","SI",
                "DLE","DC1","DC2","DC3","DC4","NAK","SYN","ETB",
                "CAN","EM","SUB","ESC","FS","GS","RS","US","SP"][c]
    elif c == 0x7F:
        return "DEL"
    return ""
```

```python
header1 = "Dec  Bin       Oct  Hex CHR"
header2 = "Dec  Bin       Oct  Hex"
def i2l(i):
    return f"{i:3d}  {i:08b}  {i:03o}  {i:02X}  {c2a(i)}"
def i2l2(i):
    return f"{i:3d}  {i:08b}  {i:03o}  {i:02X}"
print(f"{header1} | {header1} | {header2}| {header2}")
for i in range(64):
    print(f"{i2l(i):<27} | {i2l(i+64):<27} | {i2l2(i+128):<23}| {i2l2(i+192):<23}")
```
