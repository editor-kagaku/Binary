```python
def cat(filepath):
    with open(filepath) as f:
        body = f.read()
        print(body)
```

```python
cat("sample_data/README.md")
```

```python
cat("/usr/bin/ls")
# テキストモードでファイルを開いているので実行エラー
```

```python
def catb(filepath):
    with open(filepath,'rb') as f:
        body = f.read()
        print(body)
```

```python
catb("sample_data/README.md")
```

```python
catb("/usr/bin/ls")
```
