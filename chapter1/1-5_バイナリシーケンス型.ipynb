```python
a = b"This is a bytes object"
print(a)
b = bytes(10)
print(b)
c = bytes([1,2,3])
print(c)
d = bytes.fromhex("01234567 89abcdef")
print(d)
e = "バイナリ解析".encode()
print(e)
```

```python
print(a.hex())
print(a.hex(" "))
```

```python
a = bytearray(b"This is a bytearray object")
print(a)
b = bytearray(10)
print(b)
c = bytearray([1,2,3])
print(c)
d = bytearray.fromhex("01234567 89abcdef")
print(d)
```

```python
print(a.hex())
print(a.hex(" "))
```

```python
a = "バイナリ解析".encode()
print(a)
b = a.decode()
print(b)
```

```python
a = "バイナリ解析".encode("sjis")
print(a)
print(a.decode(errors="ignore"))
print(a.decode(errors="replace"))
print(a.decode(errors="backslashreplace"))
```

```python
a = b"This is a bytes object."
f = b"is"
print(f in a)
print(a.find(f))
print(a.index(f))
print(a.rfind(f))
print(a.rindex(f))
```

```python
#バイナリデータの検索(最後の行はエラー)
a = b"This is a bytes object."
f = b"Byte"
print(f in a)
print(a.find(f))
print(a.index(f))
```

```python
a = b"""for i in range(10):
    print(i)"""
print(a.split())
print(a.split(b" "))
print(a.split(b" ",1))
print(a.rsplit(b" ",1))
```
