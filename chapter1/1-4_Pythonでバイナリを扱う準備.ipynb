```python
num = [0b10, 0o10, 0x10]
print(num)
```

```python
num = 123
print(bin(num))
print(oct(num))
print(hex(num))
```

```python
print(bin(num)[2:])
print(oct(num)[2:])
print(hex(num)[2:])
```

```python
print("{:08b}".format(num))
print(f"{num:08b}")
```

```python
print("10進数\t2進数\t8進数\t16進数")
for i in range(17):
    print(f"{i:d}\t{i:04b}\t{i:02o}\t{i:X}")
```

```python
print(int("11",2))
print(int("11",8))
print(int("11",16))
```

```python
print(int("0b11",0))
print(int("0o11",0))
print(int("0x11",0))
```

```python
a=13
print(f"{a:08b}",f"{a&0xff:08b}",a,a&0xff)

a=-13
print(f"{a:08b}",f"{a&0xff:08b}",a,a&0xff)
```

```python
a = 13
print("a    :",a,f"{a:08b}")
b = a << 1
print("a<<1 :",b,f"{b:08b}")
c = a >> 1
print("a>>1 :",c,f"{c:08b}")
```

```python
a = -13
print("a    :",a,f"{a:08b}",f"{a&0xff:08b}")
b = a << 1
print("a<<1 :",b,f"{b:08b}",f"{b&0xff:08b}")
c = a >> 1
print("a>>1 :",c,f"{c:08b}",f"{c&0xff:08b}")
```

```python
a = -13&0xff
print("a    :",a,f"{a:08b}")
b = (a << 1)&0xff
print("a<<1 :",b,f"{b:08b}")
c = (a >> 1)&0xff
print("a>>1 :",c,f"{c:08b}")
```
