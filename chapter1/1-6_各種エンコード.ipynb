```python
import codecs

a = b"This is a bytes object."
a = codecs.encode(a.decode(),"rot13").encode()
print(a)
a = codecs.encode(a.decode(),"rot13").encode()
print(a)
```

```python
import base64

a = b"This is a bytes object."
a = base64.b64encode(a)
print(a)
a = base64.b64decode(a)
print(a)
```

```python
import zlib

a = b"This is a bytes object."
a = zlib.compress(a)
print(a)
a = zlib.decompress(a)
print(a)
```
